### Limpieza de Columna Edad (Age)
- Relleno de nulos con una heuristica del titulo del nombre del pasajero

    ~ Mr o Mrs -> Un rango de edad entre 35 y 85. Cualquier otro titulo lo asignamos como 'Jovenes' con un rango entre 20 y 32

### Creacion de Pipeline y Transformador de Columnas para el entrenamiento del modelo

#### Separacion de la data

```python
y = df['Survived']
X = df.drop('Survived', axis=1)
```

#### Column Transformer

Importamos Metodos de Preprocesamiento de la Data

- One Hot Encoder
- Target Enconder
- Ordinal Encoder

##### Seleccion manual de Metodo de Preprocesamiento segun la cantidad de valores unicos que posee

- Ordinal Encoder = Psclass nos dice la clase del boleto primera, segunda y tercera en esa misma jerarquia.
- One Hot Encoder = Embarked al tener pocas columnas utilizamos One Hot Encoder
- Targer Encoder = Para columnas que posean mucho valores unicos: Cabin, Name, Ticket

```python
tran = ColumnTransformer(transformers=[('columnas_con_gerarquia', OrdinalEncoder(categories=[['3','2','1']]), ['Pclass']),
                                                  ('columnas_poca_categorias', OneHotEncoder(handle_unknown='ignore'), ['Embarked']), 
                                                  ('columnas_muchas_categorias', TargetEncoder(), ['Cabin', 'Name', 'Ticket'])], remainder='passthrough',
                                                    verbose=True)
```

#### Declaracion y Configuracion de Parametros del Modelo (Random Forest Clasifier)

```python
model = RandomForestClassifier(n_estimators=130, verbose=1, random_state=42)
```

#### Creacion De Pipeline para mejor Flujo de Trabajo Preprocesamietno -> Entrenamiento

```python
pipeline = Pipeline(steps=[('Preprocesamiento', tran),
                                       ('Modelo', model)],
                                        verbose=True)
```

#### Separacio de la Data Original para Testeo del modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
X_test
```

#### Entrenamiento del Modelo

```python
pipeline.fit(X_train,y_train)
```

#### Prediccion con el 20% de la data

```python
y_predict_data_original = pipeline.predict(X_test)
```

```python
importancias_variables = pipeline.named_steps['Modelo'].feature_importances_
nombre_columnas = pipeline.named_steps['Preprocesamiento'].get_feature_names_out()
```

```python
df_importancias_var = pd.DataFrame({'Variable': nombre_columnas,
                                    'Importancia': np.round(importancias_variables, 3)}).sort_values(by='Importancia',ascending=True)
```

```python
df_importancias_var
```

```python
fig, ax = plt.subplots(figsize=(12,6))
colores = plt.get_cmap('viridis')
posicion = np.linspace(0,1, len(df_importancias_var['Variable']))
barra = ax.barh(df_importancias_var['Variable'],df_importancias_var['Importancia'], color=colores(posicion))
ax.bar_label(barra)
ax.set_xlim(0,0.5)
ax.set_title('Importancia de las Columnas Utilizadas en el Entrenamiento del Modelo (Random Forest Clasifier)')
plt.tick_params(axis='x', labelrotation=45)
plt.tight_layout()
plt.show()
```

###### Obsevamos que tenemos 6 variables que poseen menos de un 5% de importancia en la prediccion del modelo, que vendrian representando el 46% total de las columnas utilizadas, en un caso de mejoria de este modelo se podria reducir a unicamente las que poseean en este caso un valor igual o similar a un 7%

##### Conteo de la prediccion

```python
Counter(y_predict_data_original)
```

```python
ac = accuracy_score(y_test,y_predict_data_original)
precision = precision_score(y_test, y_predict_data_original)
recall = recall_score(y_test, y_predict_data_original)
metrics_list = [ac,precision,recall]
```

##### - Observamos que la metrica accuracy esta por encima del 80%, lo cual indica la buena precision que posee el modelo a la hora de clasificar.

##### - Precision igualmente por encima del 85% bastante optima para este contexto ya que indica poca aparicion de falsos positivos lo que nos dice que en este caso cuando el modelo indicaba que un tripulante no sobrevivio pero realmente este si sobrevivio.
